### Limancon of Pascal

Limancon of Pascal is a parametric curve.

<!-- ![SegmentLocal](limancon.gif "segment") -->

```python
# Limancon of Pascal
a = 1
b = 5

# symbolic variable
t = sp.Symbol("t")

# equaitions of limancon
l0 = sp.Poly((1 + t**2) ** 2, t, domain="RR")
l1 = sp.Poly(b * (1 - t**2) * (1 + t**2) + a * (1 - t**2) ** 2, t, domain="RR")
l2 = sp.Poly(2 * b * t * (1 + t**2) + 2 * a * t * (1 - t**2), t, domain="RR")

# initialize limancon
limancon = RationalCurve([l0, l1, l2, l0])

limancon.set_of_polynomials
```

```python
# plot
ax = limancon.plot()
```
